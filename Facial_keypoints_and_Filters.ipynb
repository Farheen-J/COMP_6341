```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.image as mpimg
import pandas as pd
import os
import cv2
%matplotlib inline
```

```python
image = cv2.imread('images/mona_lisa.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# plotting the image
fig = plt.figure(figsize=(9,9))
plt.imshow(image)
```

```python
#haarcascade Classifier
face_cascade = cv2.CascadeClassifier('detector_architectures/haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(image, 1.2, 2)
image_with_detections = image.copy()

# looping over the detected faces and marking the image where face is found
for (x,y,w,h) in faces:
    cv2.rectangle(image_with_detections,(x,y),(x+w,y+h),(255,0,0),3) 

fig = plt.figure(figsize=(9,9))

plt.imshow(image_with_detections)
```

```python
import torch
from models import Net

net = Net()
net.load_state_dict(torch.load('saved_models/keypoints_model_1.pt'))
net.eval()
```

```python
def show_all_keypoints(image, predicted_keypts, gt_pts=None):
    # image in grayscale
    plt.imshow(image, cmap='gray')
    plt.scatter(predicted_keypts[:, 0], predicted_keypts[:, 1], s=20, marker='.', c='m')
    # plotting ground truth points as green pts
    if gt_pts is not None:
        plt.scatter(gt_pts[:, 0], gt_pts[:, 1], s=20, marker='.', c='g')
```

```python
def visualize_output(image_test, predicted_keypts_test, gt_pts=None, batch_size=10):        
    plt.figure(figsize=(20,10))
    image=image_test[0].data
    # un-transform the image data
    image = image.numpy()   # converting to numpy array from a Tensor
    print(image.shape)
    image = np.transpose(image, (1, 2, 0)) 
    predicted_keypts =predicted_keypts_test[0].data
    predicted_keypts = predicted_keypts.numpy()
    # undo normalization of keypoints  
    predicted_keypts = predicted_keypts*50+100  
    # call show_all_keypoints
    show_all_keypoints(np.squeeze(image), predicted_keypts)
    plt.show()
```

```python
image_copy = np.copy(image)

# looping over the detected faces 
for (x,y,w,h) in faces:
    roi = image_copy[y:y+h, x:x+w]
    roi=cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)
    roi=roi/255.0
    roi = cv2.resize(roi, (224, 224))
    roi = roi.reshape(roi.shape[0], roi.shape[1], 1)
    roi = roi.transpose((2, 0, 1))
    roiT=torch.from_numpy(roi)
    roiT = roiT.type(torch.FloatTensor)
    roiT.unsqueeze_(0)
    output_pts = net(roiT)
    output_pts = output_pts.view(output_pts.size()[0], 68, -1)        
    visualize_output(roiT, output_pts)    
```

```python
# loading training data
keypoints_frame = pd.read_csv('data/training_frames_keypoints.csv')
print('Number of images: ', keypoints_frame.shape[0])
```

```python
# helper function to display keypoints
def show_keypoints(image, keypts):
    plt.imshow(image)
    plt.scatter(keypts[:, 0], keypts[:, 1], s=20, marker='.', c='m')
```

```python
#selected image
n = 130
image_name = keypoints_frame.iloc[n, 0]
image = mpimg.imread(os.path.join('data/training/', image_name))
keypts = keypoints_frame.iloc[n, 1:].to_numpy()
keypts = keypts.astype('float').reshape(-1, 2)

print('Image name: ', image_name)

plt.figure(figsize=(5, 5))
show_keypoints(image, keypts)
plt.show()
```

```python
# Display sunglasses on top of the image 
image_copy = np.copy(image)
x = int(keypts[17, 0])
y = int(keypts[17, 1])
h = int(abs(keypts[27,1] - keypts[34,1]))
w = int(abs(keypts[17,0] - keypts[26,0]))
sunglasses = cv2.imread('images/sunglasses.png', cv2.IMREAD_UNCHANGED)
new_sunglasses =  cv2.resize(sunglasses, (w, h), interpolation = cv2.INTER_CUBIC)
roi_color = image_copy[y:y+h,x:x+w]
ind = np.argwhere(new_sunglasses[:,:,3] > 0)
for i in range(3):
    roi_color[ind[:,0],ind[:,1],i] = new_sunglasses[ind[:,0],ind[:,1],i]    
image_copy[y:y+h,x:x+w] = roi_color
plt.imshow(image_copy)
```
